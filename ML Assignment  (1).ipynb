1. What is a parameter?
   
A parameter in Machine Learning is a variable that is learned from the data during the training process and determines how the model makes predictions. Parameters are internal to the model and are adjusted automatically to minimize the error between predicted and actual values.
For example, in linear regression (y = mx + c), the slope (m) and intercept (c) are parameters. The learning algorithm updates these values using training data to achieve the best fit line. Proper tuning of parameters is essential because they directly impact the model’s accuracy and performance.


2. What is correlation? What does negative correlation mean?
   
Correlation is a statistical measure that indicates the degree to which two variables are related. It shows both the strength and direction of   the relationship between variables.
The correlation coefficient ranges from -1 to +1:
+1 → Perfect positive correlation
0 → No correlation
-1 → Perfect negative correlation
Correlation helps in understanding dependencies between variables and is widely used in feature selection in Machine Learning

A negative correlation occurs when two variables move in opposite directions. This means that as one variable increases, the other decreases.
For example, the relationship between price and demand is usually negative. When the price of a product increases, the demand tends to decrease. Negative correlation is important in identifying inverse relationships in datasets.
   

3. Define Machine Learning. What are the main components in Machine Learning?

Machine Learning (ML) is a subset of Artificial Intelligence that enables systems to learn patterns from data and make decisions or predictions without being explicitly programmed.
Main Components:
Data: Raw information used for training
Features: Important attributes or variables
Model: Algorithm that learns patterns (e.g., regression, classification)
Loss Function: Measures prediction error
Optimizer: Minimizes loss by updating parameters
Evaluation Metrics: Measures model performance (accuracy, precision, recall)
These components work together to build an effective ML system

4. How does loss value help in determining whether the model is good or not?

The loss value measures how well a model’s predictions match the actual data. It quantifies the error between predicted and true values.
A low loss value indicates that the model is making accurate predictions.
A high loss value indicates poor predictions.
During training, the goal of the model is to minimize the loss function using optimization techniques. Monitoring loss also helps detect problems like overfitting or underfitting.

5. What are continuous and categorical variables?

Continuous variables are numerical variables that can take any value within a range. Examples include height, weight, temperature, and income.
Categorical variables represent categories or groups and do not have numerical meaning. Examples include gender, color, city, etc.
Understanding the type of variable is important because different preprocessing techniques are applied to each type in Machine Learning

6. How do we handle categorical variables in Machine Learning? What are the common techniques?

Machine Learning models require numerical input, so categorical variables must be converted into numbers.
Common Techniques:
Label Encoding: Assigns a unique number to each category
One-Hot Encoding: Creates binary columns for each category
Ordinal Encoding: Used when categories have a meaningful order
Choosing the right encoding method is important to avoid misleading the model.

7.What do you mean by training and testing a dataset

In Machine Learning, the available dataset is divided into two main parts: training dataset and testing dataset.
The training dataset is used to train the model. The model learns patterns, relationships, and parameters from this data.
The testing dataset is used to evaluate the model’s performance on unseen data.
This separation is important because if we test the model on the same data it was trained on, it may give very high accuracy but fail on new data. This problem is called overfitting.
Sometimes, a third dataset called the validation set is also used to tune the model during training

8. What is sklearn.preprocessing?

Scikit learn preprocessing is a module that provides functions to prepare raw data before feeding it into a Machine Learning model.
It includes:
Scaling techniques (Standardization, Normalization)
Encoding methods (Label Encoding, One-Hot Encoding)

Preprocessing is necessary because raw data is often unclean, inconsistent, or in a format that ML models cannot directly use. Proper preprocessing improves model accuracy and efficiency

9. What is a Test set?

A test set is a portion of the dataset that is kept completely separate from the training process and is used only to evaluate the final model.
Key points:
It contains unseen data
It provides an unbiased evaluation
It helps check whether the model can generalize well
A good model should perform well not only on training data but also on the test set.

12. What is correlation?

Correlation is a statistical technique used to measure the strength and direction of the relationship between two variables. It helps in understanding how one variable changes with respect to another.
The value of correlation is expressed using the correlation coefficient (r), which ranges from -1 to +1:
+1 → Perfect positive correlation
0 → No correlation
-1 → Perfect negative correlation
Correlation is widely used in Machine Learning for feature selection, as highly correlated features may be redundant and can affect model performance.


13. What does Negative Correlation Mean?

A negative correlation indicates an inverse relationship between two variables. This means:
When one variable increases, the other decreases
When one decreases, the other increases
For example, there is usually a negative correlation between speed and time (for a fixed distance). If speed increases, the time taken decreases.
Negative correlation is useful in identifying opposing trends in data.

15. What is causation? Explain difference between correlation and causation with an example

Causation means that one variable directly influences or causes a change in another variable.
Difference:
Correlation: Indicates a relationship between two variables but does not imply cause
Causation: Indicates a direct cause-and-effect relationship
Example:
Causation: Lack of sleep → causes tiredness
Correlation: Ice cream sales and drowning cases both increase in summer, but one does not cause the other
Understanding this difference is important because relying only on correlation can lead to incorrect conclusions in Machine Learning and data analysis.


16. What is an Optimizer? What are different types of optimizers? Explain each with an example.

An optimizer is an algorithm used in Machine Learning to minimize the loss function by updating the model’s parameters during training.
It works by adjusting weights and biases in such a way that the prediction error is reduced.
Types of Optimizers:
Gradient Descent: Uses the entire dataset to update parameters
Stochastic Gradient Descent (SGD): Updates parameters for each data point
Mini-batch Gradient Descent: Uses small batches of data
Adam (Adaptive Moment Estimation): Combines momentum and adaptive learning rate; widely used
RMSprop: Adjusts learning rate based on recent gradients
Optimizers are essential for improving model accuracy and ensuring faster convergence.

21. What is Feature Scaling? How does it help in Machine Learning??

Feature scaling is the process of bringing all feature values to a similar scale or range.
Importance:
Prevents one feature from dominating others
Improves performance of algorithms like KNN, SVM
Helps gradient descent converge faster
Leads to better accuracy
Common scaling methods include Standardization and Normalization.
