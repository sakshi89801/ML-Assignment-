 1. What is a Support Vector Machine (SVM)

 Support Vector Machine (SVM) is a powerful supervised machine learning algorithm used for both classification and regression tasks. The main objective of SVM is to find the optimal hyperplane that separates data points of different classes with the maximum possible margin.

In a 2D space, the hyperplane is simply a line, but in higher dimensions, it becomes a plane or hyperplane. SVM focuses on those data points that are closest to the decision boundary, called support vectors, as they are critical in defining the position of the hyperplane.

SVM is highly effective in:

High-dimensional spaces
Cases where number of features > number of samples
Non-linear classification using kernels

2. Difference between Hard Margin and Soft Margin SVM

Hard Margin SVM assumes that the data is perfectly linearly separable. It tries to find a boundary that separates classes without any error. However, this is rarely practical in real-world datasets, which often contain noise.

Soft Margin SVM allows some misclassification by introducing slack variables. It balances between maximizing margin and minimizing classification error.

Hard Margin:
No misclassification allowed
Sensitive to noise
Overfitting risk
Soft Margin:
Allows errors
More robust
Controlled using parameter C

Thus, Soft Margin SVM is widely used in practical applications.

4. Role of Lagrange Multipliers in SVM

Lagrange multipliers are used to solve the constrained optimization problem of SVM. Instead of solving the problem directly, it is converted into a dual optimization problem using Lagrangian formulation.

Advantages:

Makes computation efficient
Allows use of kernel functions
Focuses only on support vectors

The dual form depends only on dot products of data points, which is crucial for applying the kernel trick.

5. What are Support Vectors?

Support vectors are the critical data points that lie closest to the decision boundary (hyperplane). These points directly influence the position and orientation of the hyperplane.

Key properties:

Only support vectors affect the model
Removing other points does not change boundary
They define the margin

Thus, SVM is memory efficient as it uses only a subset of training data.

6. What is Support Vector Classifier (SVC)?

Support Vector Classifier (SVC) is the classification implementation of SVM. It is used to separate data into different categories by finding an optimal hyperplane.

It can handle:

Binary classification
Multi-class classification (using OvR or OvO strategies)

SVC also supports different kernels such as linear, polynomial, and RBF.

7. What is Support Vector Regressor (SVR)?

Support Vector Regressor (SVR) is used for regression tasks. Instead of minimizing classification error, SVR tries to fit a function within a margin of tolerance called epsilon (ε).

Key idea:

Errors within ε are ignored
Only large deviations are penalized

This makes SVR robust to outliers.

8. What is Kernel Trick?

Kernel trick is a technique used in SVM to handle non-linear data. Instead of explicitly transforming data into higher dimensions, kernel functions compute the dot product in that space directly.

This avoids heavy computation and allows SVM to learn complex patterns efficiently.

9. Comparison of Kernels
Linear Kernel
Used when data is linearly separable. Fast and simple.
Polynomial Kernel
Captures interactions between features. Useful for curved decision boundaries.
RBF Kernel (Gaussian)
Most powerful and widely used. Handles highly complex relationships by mapping data to infinite dimensions.

10. Effect of C Parameter

The parameter C controls the trade-off between:

Maximizing margin
Minimizing classification error
Small C:
Large margin
More misclassification
Better generalization
Large C:
Small margin
Less error
Risk of overfitting

11. Role of Gamma in RBF Kernel
Gamma determines how far the influence of a single training point reaches.
Low gamma:
Large influence
Smooth boundary
High gamma:
Small influence
Complex boundary
Proper tuning of gamma is essential to avoid underfitting or overfitting.


12. What is Naïve Bayes?
Naïve Bayes is a probabilistic classifier based on Bayes’ Theorem. It calculates the probability of a class given the input features and assigns the class with the highest probability.

It is widely used in:
Spam detection
Text classification
Sentiment analysis

14. Why is it called “Naïve”?
It is called “naïve” because it assumes that all features are independent of each other, which is rarely true in real-world data. Despite this simplification, it performs surprisingly well.

15. Types of Naïve Bayes
Gaussian NB: For continuous data assuming normal distribution
Multinomial NB: For count data (e.g., word frequency)
Bernoulli NB: For binary features

16. When to use Gaussian NB?
Use Gaussian NB when:
Features are continuous
Data follows normal distribution
No strong correlations between features

17. Assumptions of Naïve Bayes
Feature independence
Equal contribution of features
Correct distribution assumption (Gaussian for continuous data)
These assumptions simplify computation significantly.

18. Advantages and Disadvantages
Advantages:
Fast and efficient
Works well with small datasets
Handles high-dimensional data
Easy to implement
Disadvantages:
Unrealistic independence assumption
Poor performance on correlated features
Probability estimates may not be accurate


19. Why Naïve Bayes is good for Text Classification?

Naïve Bayes works extremely well in text classification because:
Text data is high-dimensional
Word occurrences are treated independently
Works well with sparse matrices
Hence, it is widely used in spam filtering and document classification.

20. SVM vs Naïve Bayes

SVM is a discriminative model that focuses on finding decision boundaries, while Naïve Bayes is a generative model that models probability distributions.
SVM:
High accuracy
Works well with complex boundaries
Slower
Naïve Bayes:
Faster
Works well with text data
Less accurate in complex cases

```python
#21
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

iris = datasets.load_iris()
X, y = iris.data, iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = SVC(kernel='linear')
model.fit(X_train, y_train)

print("Accuracy:", accuracy_score(y_test, model.predict(X_test)))
```

```python
#24
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC

iris = datasets.load_iris()
X = iris.data[:, :2]
y = iris.target

model = SVC(kernel='poly', degree=3)
model.fit(X, y)

# Plot
x_min, x_max = X[:,0].min()-1, X[:,0].max()+1
y_min, y_max = X[:,1].min()-1, X[:,1].max()+1

xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))

Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z)
plt.scatter(X[:,0], X[:,1], c=y)
plt.show()
```

```python
#35
from sklearn.datasets import load_breast_cancer
from sklearn.feature_selection import SelectKBest
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB

data = load_breast_cancer()
X, y = data.data, data.target

# Without feature selection
X_train, X_test, y_train, y_test = train_test_split(X, y)
model1 = GaussianNB()
model1.fit(X_train, y_train)
print("Without FS:", model1.score(X_test, y_test))

# With feature selection
selector = SelectKBest(k=10)
X_new = selector.fit_transform(X, y)

X_train, X_test, y_train, y_test = train_test_split(X_new, y)
model2 = GaussianNB()
model2.fit(X_train, y_train)
print("With FS:", model2.score(X_test, y_test))
```

```python
#39
from sklearn.naive_bayes import GaussianNB
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X, y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y)

priors_list = [[0.3, 0.3, 0.4], [0.2, 0.5, 0.3]]

for p in priors_list:
    model = GaussianNB(priors=p)
    model.fit(X_train, y_train)
    print("Priors:", p, "Accuracy:", accuracy_score(y_test, model.predict(X_test)))
```

```python
#40
from sklearn.feature_selection import RFE
from sklearn.svm import SVC
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X, y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y)

# Without RFE
model = SVC(kernel='linear')
model.fit(X_train, y_train)
print("Without RFE:", accuracy_score(y_test, model.predict(X_test)))

# With RFE
rfe = RFE(model, n_features_to_select=2)
X_train_rfe = rfe.fit_transform(X_train, y_train)
X_test_rfe = rfe.transform(X_test)

model.fit(X_train_rfe, y_train)
print("With RFE:", accuracy_score(y_test, model.predict(X_test_rfe)))
```
