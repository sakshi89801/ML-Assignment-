2: What are the key assumptions of Simple Linear Regression?

Answer:
Simple Linear Regression relies on several important assumptions to produce valid results:

Linearity
The relationship between independent and dependent variables must be linear.
Independence of Errors
The residuals (errors) should be independent of each other, meaning one observation does not influence another.
Homoscedasticity
The variance of errors should be constant across all levels of the independent variable.
Normality of Errors
The residuals should be normally distributed, especially important for hypothesis testing.
No Multicollinearity
(More relevant in multiple regression) Independent variables should not be highly correlated.
No Autocorrelation
Errors should not be correlated over time (important in time-series data).

If these assumptions are violated, the model may give biased or inefficient results.

3: What is heteroscedasticity, and why is it important to address in regression models?

Answer:
Heteroscedasticity refers to a condition where the variance of the residuals (errors) is not constant across all levels of the independent variable.

In simple terms, the spread of errors increases or decreases as the value of the predictor changes.

Why it is important:

It violates one of the key assumptions of linear regression (homoscedasticity).
It leads to inefficient estimates of coefficients.
Standard errors become unreliable.
Hypothesis tests (like t-tests and F-tests) may give incorrect results.
Confidence intervals may be misleading.

Example:
If predicting income based on experience, variability in income may increase at higher experience levels, causing heteroscedasticity.

How to detect and fix:

Residual plots
Breusch-Pagan test
Use transformations (log, square root)
Apply weighted least squares

Addressing heteroscedasticity improves model accuracy and reliability.

```python
#6: Simple Linear Regression Implementation in Python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Given data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])

# Create model
model = LinearRegression()

# Train model
model.fit(X, Y)

# Predictions
Y_pred = model.predict(X)

# Print coefficients
print("Intercept (b0):", model.intercept_)
print("Slope (b1):", model.coef_[0])

# Plotting
plt.scatter(X, Y, label="Data Points")
plt.plot(X, Y_pred, label="Regression Line")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Simple Linear Regression")
plt.legend()
plt.show()
```

```python
#7: Multiple Linear Regression with VIF Check
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Given data
data = {
    "Area": [1200, 1500, 1800, 2000],
    "Rooms": [2, 3, 3, 4],
    "Price": [250000, 300000, 320000, 370000]
}

df = pd.DataFrame(data)

# Independent and dependent variables
X = df[["Area", "Rooms"]]
Y = df["Price"]

# Fit Multiple Linear Regression model
model = LinearRegression()
model.fit(X, Y)

# Predictions
Y_pred = model.predict(X)

# Print coefficients
print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)

# -------- VIF Calculation --------
vif_data = pd.DataFrame()
vif_data["Feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

print("\nVIF Values:")
print(vif_data)
```

```python
# 8: Polynomial Regression
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Given data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

# Convert to polynomial features (degree 2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Train model
model = LinearRegression()
model.fit(X_poly, Y)

# Predictions
Y_pred = model.predict(X_poly)

# Print coefficients
print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)

# Plotting curve (smooth line)
X_range = np.linspace(1, 5, 100).reshape(-1, 1)
X_range_poly = poly.transform(X_range)
Y_range_pred = model.predict(X_range_poly)

# Plot
plt.scatter(X, Y, label="Data Points")
plt.plot(X_range, Y_range_pred, label="Polynomial Curve")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Polynomial Regression (Degree 2)")
plt.legend()
plt.show()
```

```python
#9: Residual Plot & Heteroscedasticity Check
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Given data
X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
Y = np.array([15, 35, 40, 50, 65])

# Train model
model = LinearRegression()
model.fit(X, Y)

# Predictions
Y_pred = model.predict(X)

# Residuals
residuals = Y - Y_pred

# Print values
print("Predicted Values:", Y_pred)
print("Residuals:", residuals)

# Residual plot
plt.scatter(X, residuals)
plt.axhline(y=0)  # horizontal line at 0
plt.xlabel("X")
plt.ylabel("Residuals")
plt.title("Residual Plot")
plt.show()
```

10: Handling Heteroscedasticity and Multicollinearity in a Regression Model

Answer:

As a data scientist working for a real estate company, predicting house prices accurately is very important. When building a regression model using features such as area, number of rooms, and location, two common issues that may arise are heteroscedasticity and multicollinearity. These issues can reduce the reliability and interpretability of the model. To ensure a robust model, the following steps can be taken.

1. Detect the Problems

Before fixing the issues, the first step is identifying them.

For Heteroscedasticity:

Plot residual vs predicted values to check if the variance of residuals changes.
Perform statistical tests such as the Breusch–Pagan test or White test.
If the residuals fan out or form patterns, heteroscedasticity is present.

For Multicollinearity:

Calculate Variance Inflation Factor (VIF) for each independent variable.
If VIF > 5, it indicates moderate multicollinearity.
If VIF > 10, it indicates severe multicollinearity.
Also check the correlation matrix among features.
2. Address Heteroscedasticity

If heteroscedasticity is detected, several techniques can be used:

a. Transform the Target Variable

Apply transformations such as log, square root, or Box–Cox transformation to stabilize variance.
Example: Using log(price) instead of price.

b. Weighted Least Squares (WLS)

Assign different weights to observations so that points with larger variance have less influence on the model.

c. Robust Regression Methods

Use models that are less sensitive to heteroscedasticity, such as Huber regression or quantile regression.
3. Address Multicollinearity

To reduce multicollinearity between features like area and number of rooms:

a. Remove Highly Correlated Features

If two variables are highly correlated, one of them can be removed.

b. Feature Engineering

Combine correlated variables into a single meaningful feature (for example, price per square foot).

c. Regularization Techniques

Apply Ridge Regression to reduce coefficient variance.
Use Lasso Regression to perform feature selection automatically.

d. Principal Component Analysis (PCA)

Convert correlated variables into a smaller number of uncorrelated components.
4. Model Validation

After fixing the issues:

Re-train the regression model.
Check residual plots again to ensure variance is constant.
Evaluate performance using metrics such as R², RMSE, and MAE.
Use cross-validation to confirm the model generalizes well to new data.
5. Final Deployment

Once the model is stable and validated:

Use the cleaned and optimized model for predicting house prices.
Continuously monitor the model as new housing data becomes available.
